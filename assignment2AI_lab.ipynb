```python
from sklearn.datasets import load_digits 
import matplotlib.pyplot as plt
#load the dataset
digits = load_digits()
print("dataset shape:",digits.data.shape)
print("image shape:",digits.images.shape)
print("target labels:",digits.target_names)
plt.figure(figsize=(4,4))
plt.imshow(digits.images[4],cmap='pink')
plt.title(f"label:{digits.target[2]}")
plt.axis('on')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load the dataset
digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='pink')
    ax.set_title(f"Label: {digits.target[i]}")
    ax.axis('off')   # Hide axes

plt.tight_layout()
plt.show()

```

```python
#import libraries 
import pandas as pd
from sklearn.datasets import load_iris

#load the irirs dataset
iris = load_iris( as_frame=True)
df = iris.frame
print("number of rows and column")
print(df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
      
```
